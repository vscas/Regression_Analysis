#### Realizando a regressÃ£o

from sklearn.linear_model import LinearRegression

```python
from scipy import stats
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import StandardScaler
```

```python
model = LinearRegression()
scaler = StandardScaler()
```

```python
X1_1 = df['Posse de Bola']
X1_5 = df['Faltas Cometidas']
X1_6 = df['Escanteio']
X1_7 = df['Impedimento']
X1_8 = df['Chute ao Gol']
X1_9 = df['Passes Certos']
X1_11 = df['Defesas']
X1_12 = df['Faltas sofridas']
X1_13 = df['Chutes Errados']
X1_14 = df['Divididas Ganhas']
```

#### Posse de Bola

```python
x1 = df[['Gols Feitos','Gols Sofridos','Total de Chutes','Faltas Cometidas','Escanteio','Impedimento','Chute ao Gol','Passes Certos','Total de Passes','Defesas','Faltas sofridas','Chutes Errados','Divididas Ganhas']]

y1 = df['Posse de Bola']

# normalizando os dados
scaler.fit(x1)
x1_std = scaler.transform(x1)

# carregando o modelo
model.fit(x1_std, y1)

# retorna a
a1 = model.coef_

# retorna b
b1 = model.intercept_

# retorna RÂ²
r2_1 = model.score(x1_std, y1)

# retorna o erro
y_p1=model.predict(x1_std)
e1=np.sqrt(mean_squared_error(y1, y_p1))

print(f''' Posse de Bola

Constante = {b1}
RÂ² = {r2_1}
RMSE = {e1}
''')
```

```python
slope, intercept, r_value, p_value_x1_1, std_err = stats.linregress(X1_2, X1_1)
slope, intercept, r_value, p_value_x1_5, std_err = stats.linregress(X1_6, X1_1)
slope, intercept, r_value, p_value_x1_6, std_err = stats.linregress(X1_7, X1_1)
slope, intercept, r_value, p_value_x1_7, std_err = stats.linregress(X1_8, X1_1)
slope, intercept, r_value, p_value_x1_8, std_err = stats.linregress(X1_9, X1_1)
slope, intercept, r_value, p_value_x1_9, std_err = stats.linregress(X1_10, X1_1)
slope, intercept, r_value, p_value_x1_11, std_err = stats.linregress(X1_12, X1_1)
slope, intercept, r_value, p_value_x1_12, std_err = stats.linregress(X1_13, X1_1)
slope, intercept, r_value, p_value_x1_13, std_err = stats.linregress(X1_14, X1_1)

p_value_posse = [p_value_x1_1,p_value_x1_5,p_value_x1_6,p_value_x1_7,p_value_x1_8,p_value_x1_9,p_value_x1_11,p_value_x1_12,p_value_x1_13]
```

```python
#correlaÃ§Ã£o das variaveis
corr1 = pd.DataFrame({'Columns':x1.columns,'coefs':model.coef_, 'p value':p_value_posse})
corr1.round(4)
```
